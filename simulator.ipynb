# Simulation of Cluster Formation on Regular Graphs

## Graph Building

```python
import random
import numpy as np
from tqdm import tqdm



def indicator(p):
    return 1 if random.random() < p else 0

class Edge:
    def __init__(self, p):
        self.open = indicator(p)
        self.children=[]
        
    def connect(self, child=None):
        self.children.append(child)
        return child
    
    def __str__(self):
        return "Edge({})".format(self.open)
    
class Graph:
    def __init__(self, p, d):
        self.p = p
        self.d = d
        self.v = Edge(1)
        self.candidates = [self.v]
        self.cluster_counter = 0
        self.l = 0
        
    def add_child(self):
        parent = self.candidates.pop(0)
        children = [parent.connect(Edge(self.p)) for i in range(self.d-1)]
        children = [child for child in children if child.open == 1]
        self.candidates.extend(children)
        self.cluster_counter += len(children)
        self.l += 1
        
    def build(self, max_size=40e3):
        while self.candidates:
            self.add_child()
            if self.cluster_counter >= max_size:
                return self.cluster_counter, self.l 
        return self.cluster_counter, self.l 
       
```

```python
p = .6
d = 5
G = Graph(p, d)
print(G.build())
```

## Simulation

```python
trials = 20

lines = []

for d in range(3, 6):
    y = []
    print("d = {}".format(d))
    for p in range(0, 101):
        p = p/100
        print("\tp={}".format(str(p)))
        cluster_sizes = []
        for i in range(trials):
            G = Graph(p, d)
            cluster_size, level = G.build()
            cluster_sizes.append(cluster_size)
        y.append(np.mean(cluster_sizes))
    print("="*30)
    lines.append(y)

print(lines)
```
